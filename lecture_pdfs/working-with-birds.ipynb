````python
def makeBirdDict(filename):
    # this function opens the data file from Thomas et al and extracts the species and family as a dictionary
    birdDict = {}
    fobj = open(filename,"r", encoding = "ISO-8859-15") #open the file
    all_records = fobj.readlines()[1:] #read the file
    for line in all_records:
        record_elements = line.split(",")
        species = record_elements[0]
        family = record_elements[2]
        birdDict[species] = family
    #split lines on comma
    #stick species and family into birdDict
    
    return birdDict
````

Before we move on, lets look at a python way of searching for the family in the file

````python
import re
def grep(pattern,fileObj):
  r=[]
  for line in fileObj:
    if re.search(pattern,line):
      r.append(line)
  return r

pattern = "Cracidae"
ff = open("formatted-eBird.csv", "r")

crows = grep(pattern, ff)

print(crows)
````

lets turn the above code into a function  we can call to find all families in the 8th column no matter how they are formatted

````python
import re

def getFam(field):
    #this function returns the first element of the family field from the ebirds dataset
    pattern = r"[a-zA-Z]+\b"
    matches = (re.findall(pattern, field))
    if matches: # if we find a match
        #print ('found a match!')
        fam = matches[0]
    else:
        #print ('no match for this field')
        #print(field)
        fam = None
    return fam

getFam(crowline)
````
